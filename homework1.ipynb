```python
# 필수 1.데이터 불러오기
import pandas as pd

df = pd.read_csv("flight_data_homework.csv")

print(df.shape)

df.head()
```

```python
# 필수 2.결측치 처리
import pandas as pd

print(df.isna().sum())

df = df.dropna()

print(df.isna().sum())
```

```python
# 필수 3.조건에 맞는 데이터 추출하기
import pandas as pd

# df.groupby("Destination")["Price"].agg(["mean", "median"]).round(1)

avg_df = df.groupby("Destination")["Price"].agg(
    평균값=lambda x: round(x.mean(), 1),
    중앙값=lambda x: round(x.median(), 1)
).reset_index()

display(avg_df)

df2 = (
    df.groupby(['Airline', 'Total_Stops'])['Route']
       .nunique()
       .reset_index()
)

df2
```

```python
# 필수4. 조건에 맞는 데이터 추출하기2
import pandas as pd

pivot_df = pd.pivot_table(
    df,
    values='Airline',
    index=['Source','Destination'],
    aggfunc='count',
    fill_value=0,
)

pivot_df = pivot_df.sort_values(by='Airline', ascending=False)

display(pivot_df)

filtering_df = df[(df['Airline'] == 'Air India') & (df['Price'] >= 7000)]

display(filtering_df)
```

```python
# 도전 1. 조건에 맞는 데이터 추출하기3
import pandas as pd

df['Date_of_Journey'] = pd.to_datetime(df['Date_of_Journey'])

df['Day'] = df["Date_of_Journey"].dt.day_name()

wed_price = df[df['Day'] == 'Wednesday']['Price'].mean()

wed_price
```

```python
# 도전 2. Dep_Time 기준 비행기 시간대 분류
import pandas as pd

df['Dep_Time'] = pd.to_datetime(df['Dep_Time'])

df['Q_Day'] = df['Dep_Time'].dt.hour.apply(
    lambda h: '아침' if 5 <= h <12 else '낮' if 12 <= h < 18 else '저녁' if 18 <= h < 24 else '밤'
)

order = ['아침', '낮', '저녁', '밤']
df['Q_Day'] = pd.Categorical(df['Q_Day'], categories=order, ordered=True)

df_counts = df['Q_Day'].value_counts().reindex(order).reset_index()
df_counts.columns = ['Time', 'Counts']

display(df_counts)
```
